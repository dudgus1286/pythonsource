```python
# score.txt 읽어서 평균과 합 출력
with open("./resource/score.txt","r") as f:
    score = f.readlines()

    avg = 0
    total = 0

    for i in score:
        total += int(i)
        
    avg = total / len(score)

    print(f"점수 총합: {total} 점수 평균: {avg}")

```

```python
score=[]
# 파일 읽어온 f 그대로 for문 돌릴 수 있음
with open("./resource/score.txt","r") as f:
    for num in f:
        # print(num , end="")
        score.append(int(num))
print(sum(score))
print(sum(score)/len(score))
```

```python
# score.txt 읽어서 총합과 평균 구한 후 result.txt 작성
# result.txt 내용
# 총합 : 520
# 평균 : 86.67
with open("./resource/result.txt","w",encoding="utf-8") as f:
    score=[]
    with open("./resource/score.txt","r") as f2:
        for num in f2:
            score.append(int(num))
    f.write("총합 : %d\n평균 : %4.2f" %( sum(score), sum(score)/len(score)))
```

```python
# 500 명의 키와 몸무게 랜덤 생성 후 파일에 저장
import random
strings = list("가나다라마바사아자차카타파하")

print(random.choices(strings) + random.choices(strings)) # 이름
print(random.randrange(150,200)) # 키
print(random.randrange(50,100)) # 몸무게

with open("./resource/info.txt","w",encoding="utf-8") as f:
    for i in range(500):
        name = random.choice(strings) + random.choice(strings)
        f.write("%s, %d, %d\n" % (
            name, random.randrange(150,200), random.randrange(50,100)
            ))
```

```python
# info.txt 읽어서 BMI 지수 계산
# bmi = 몸무게 / (키 / 100 ** 2)
# bmi >= 25 과체중, bmi >= 18.5 정상체중, 그 이하 저체중

# 라라, 176, 90
# 이름 : 라라
# 키 : 176
# 몸무게 : 90
# BMI : 25
# 결과 : 정상체중
with open("./resource/info.txt","r", encoding="utf-8") as f:
    for row in f :
        name, height , weight = row.strip().split(", ")
        # print(name, height, weight)
        bmi = int(weight) / ( (int(height) / 100) ** 2)

        result=None
        if bmi >= 25:
            result = "과체중"
        elif bmi >= 18.5:
            result = "정상체중"
        else:
            result = "저체중"
        print("\n".join([
            "이름 : {}", "키 : {}", "몸무게 : {}", "BMI : {}", "결과 : {}"]).format(name,height,weight,bmi,result))
```

```python
# 변수 초기화 : "" , None, 0
content = None
while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴 번호 입력 : "))

    if no == 1:
        # origin.txt 읽은 후 content 변수에 담기
        # content 변수에 담은 값을 chr(ord(content) + 100) 변경 후 파일 작성(encryption.txt)
        with open("./resource/origin.txt","r",encoding="utf-8") as f:
            content = f.read()

            with open("./resource/encryption.txt","w",encoding="utf-8") as f:
                for c in content:
                    f.write(chr(ord(c)+100))
    elif no == 2:
        # encryption.txt 읽어온 후 원래의 문자로 출력
        with open("./resource/encryption.txt","r",encoding="utf-8") as f:
            content = f.read()
            for i in range(0, len(content)):
                print(chr(ord(content[i])-100), end="")
    else:
        break
```
