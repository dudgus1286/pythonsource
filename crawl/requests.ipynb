```python
# GET
# get Ï£ºÏÜå
url = "https://httpbin.org/get"
with requests.Session() as s :
    # get Î©îÏÜåÎìú
    response = s.get(url)
    print(response.text)
```

```python
# POST
# post Ï£ºÏÜå
url = "https://httpbin.org/post"

# "form"Ïóê ÏÇΩÏûÖÎêòÎäî Îç∞Ïù¥ÌÑ∞ dict Íµ¨Ï°∞Î°ú ÏÉùÏÑ±
data = {"name":"hong"}

with requests.Session() as s :
    # post Î©îÏÜåÎìú
    response = s.post(url, data=data)
    print(response.text)
```

```python
# PUT
# put Ï£ºÏÜå
url = "https://httpbin.org/put"

# ÌèºÏóê ÏÇΩÏûÖÎêòÎäî Îç∞Ïù¥ÌÑ∞ dict Íµ¨Ï°∞Î°ú ÏÉùÏÑ±
data = {"name":"kim"}

with requests.Session() as s :
    # put Î©îÏÜåÎìú
    response = s.put(url, data=data)
    print(response.text)
```

```python
# DELETE
# delete Ï£ºÏÜå
url = "https://httpbin.org/delete"
# Ìèº ÎÇ¥ Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú
with requests.Session() as s :
    # delete Î©îÏÜåÎìú
    response = s.delete(url)
    print(response.text)
```

```python
url = "https://httpbin.org/get"

# ÌååÎùºÎ©îÌÑ∞
# https://httpbin.org/get?name=hong ÏóêÏÑú ? Ïù¥ÌõÑ Î∂ÄÎ∂Ñ
# Í∏∞Ï°¥ urllib Ïùò Í≤ΩÏö∞ ÌååÎùºÎ©îÌÑ∞ Í∞íÏùÑ ÎîïÏÖîÎÑàÎ¶¨ ÌòïÌÉúÎ°ú ÎßåÎì† Îí§ Î≥ÑÎèÑÏùò Ïù∏ÏΩîÎî© ÏûëÏóÖÏù¥ ÌïÑÏöîÌñàÏùå
# requestsÎäî ÏûêÎèôÏúºÎ°ú ÏûëÎèô

# payload = {"name":"hong", "age":"25"}
payload = {"name":"hong", "age":"25", "key1": ["value1", "value2"]}

with requests.Session() as s:
    response=s.get(url, params=payload)
    # print(response.text)
    print(response.url)
    # Í≤∞Í≥º : https://httpbin.org/get?name=hong&age=25&key1=value1&key1=value2
    # Î≥µÏàòÏÑ†ÌÉù Í∞ÄÎä•Ìïú checkbox Ï≤òÎüº Í∞ôÏùÄ nameÏù¥ Î∞òÎ≥µÎê®
```

```python
url = "https://jsonplaceholder.typicode.com/todos/1"

with requests.Session() as s:
    response=s.get(url)
    print(response.json())
    print(response.text)
    print(response.content) # binary ÌòïÏãù
```

```python
url = "https://jsonplaceholder.typicode.com/users"

with requests.Session() as s:
    response=s.get(url)
    print(response.json())

    for row in response.json():
        for k, v in row.items():
            print(f"{k} : {v}")
```

```python
# fake_useragent Ìå®ÌÇ§ÏßÄ Ï†ÅÏö©
from fake_useragent import UserAgent
userAgent = UserAgent()
headers = {"user-agent" : userAgent.chrome}

url = "https://httpbin.org/get"

# payload = {"name":"hong", "age":"25"}
payload = {"name":"hong", "age":"25", "key1": ["value1", "value2"]}

with requests.Session() as s:
    response=s.get(url, params=payload, headers=headers)
    print(response.url)
```

```python
url = "https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=GMP&durationDays=30&count=100&_=1716515102335"
# Îã§Ïùå ÏáºÌïëÎ™∞ => Í∞úÎ∞úÏûê Ìà¥ - ÎÑ§Ìä∏ÏõåÌÅ¨ - Fetch/XHR Name Î™©Î°ù Ï§ë ÏÉÅÌíàÎ™©Î°ù json
userAgent = UserAgent()
headers = {"user-agent":userAgent.chrome}
# 1 product_name price_max

with requests.Session() as s:
    r = s.get(url, headers=headers)
    for idx, item in enumerate(r.json(),1):
        print(idx, item["product_name"], item["price_max"])
        
```
