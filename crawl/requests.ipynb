#### requests 라이브러리 사용하기
- HTTP 요청 시 사용
- urllib.request 보다 더 직관적, 간단함
- 디코딩 적절히 해줌
- json 처리 편함

```python
import requests
```

```python
response = requests.get("https://www.naver.com")

print(response.status_code)
print(response.encoding)
print(response.text)
```

```python
# 세션 활성화
s = requests.Session()
response = s.get("https://www.naver.com")
print(response.text)
s.close()
```

```python
# 세션 + with => close() 호출 안 해도 됨
with requests.Session() as s:
    response = s.get("https://www.naver.com")
    print(response.text)
```

> 다양한 http 메소드 지원   
   
(REST컨트롤러에서 썼던 것과 비슷한 것들)
- get()
- post()
- put()
- delete()   
   
연습용 사이트 ==> https://httpbin.org/

```python
# GET
# get 주소
url = "https://httpbin.org/get"
with requests.Session() as s :
    # get 메소드
    response = s.get(url)
    print(response.text)
```

```python
# POST
# post 주소
url = "https://httpbin.org/post"

# "form"에 삽입되는 데이터 dict 구조로 생성
data = {"name":"hong"}

with requests.Session() as s :
    # post 메소드
    response = s.post(url, data=data)
    print(response.text)
```

```python
# PUT
# put 주소
url = "https://httpbin.org/put"

# 폼에 삽입되는 데이터 dict 구조로 생성
data = {"name":"kim"}

with requests.Session() as s :
    # put 메소드
    response = s.put(url, data=data)
    print(response.text)
```

```python
# DELETE
# delete 주소
url = "https://httpbin.org/delete"
# 폼 내 데이터 삭제
with requests.Session() as s :
    # delete 메소드
    response = s.delete(url)
    print(response.text)
```

```python
url = "https://httpbin.org/get"

# 파라메터
# https://httpbin.org/get?name=hong 에서 ? 이후 부분
# 기존 urllib 의 경우 파라메터 값을 딕셔너리 형태로 만든 뒤 별도의 인코딩 작업이 필요했음
# requests는 자동으로 작동

# payload = {"name":"hong", "age":"25"}
payload = {"name":"hong", "age":"25", "key1": ["value1", "value2"]}

with requests.Session() as s:
    response=s.get(url, params=payload)
    # print(response.text)
    print(response.url)
    # 결과 : https://httpbin.org/get?name=hong&age=25&key1=value1&key1=value2
    # 복수선택 가능한 checkbox 처럼 같은 name이 반복됨
```

```python
url = "https://jsonplaceholder.typicode.com/todos/1"

with requests.Session() as s:
    response=s.get(url)
    print(response.json())
    print(response.text)
    print(response.content) # binary 형식
```

```python
url = "https://jsonplaceholder.typicode.com/users"

with requests.Session() as s:
    response=s.get(url)
    print(response.json())

    for row in response.json():
        for k, v in row.items():
            print(f"{k} : {v}")
```

```python
# fake_useragent 패키지 적용
from fake_useragent import UserAgent
userAgent = UserAgent()
headers = {"user-agent" : userAgent.chrome}

url = "https://httpbin.org/get"

# payload = {"name":"hong", "age":"25"}
payload = {"name":"hong", "age":"25", "key1": ["value1", "value2"]}

with requests.Session() as s:
    response=s.get(url, params=payload, headers=headers)
    print(response.url)
```

```python
url = "https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=GMP&durationDays=30&count=100&_=1716515102335"
# 다음 쇼핑몰 => 개발자 툴 - 네트워크 - Fetch/XHR Name 목록 중 상품목록 json
userAgent = UserAgent()
headers = {"user-agent":userAgent.chrome}
# 1 product_name price_max

with requests.Session() as s:
    r = s.get(url, headers=headers)
    for idx, item in enumerate(r.json(),1):
        print(idx, item["product_name"], item["price_max"])
        
```
